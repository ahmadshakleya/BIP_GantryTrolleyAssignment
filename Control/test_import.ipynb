```python
import pickle

with open('objs.pkl', 'rb') as f:  # Python 3: open(..., 'rb')
    loops = pickle.load(f)
```

```python
print(loops[0])
```

```python
# Constants
a = 33  # Weight for theta_max
b = 29  # Weight for t_task
tolerance = 0.05  # Tolerance for x_array values to approximate 20
extreme_theta_threshold = 30  # Threshold for extreme theta values

# Placeholder for results
indices_with_extreme_theta = []
costs = []

# Iterate over the list of dictionaries
for idx, loop in enumerate(loops):
    x_array = loop["x_array"]
    theta_array = loop["theta_array"]
    time_array = loop["time_array"]

    # Check if x_array contains a value approximately 20
    if any(abs(x - 20) <= tolerance for x in x_array):
        # Find the index where x_array first reaches approx 20
        index_x_approx_20 = next(i for i, x in enumerate(x_array) if abs(x - 20) <= tolerance)
        t_task = time_array[index_x_approx_20]  # Corresponding t_task

        # Check if theta_array contains extreme values (+30 or -30)
        if any(theta > extreme_theta_threshold or theta < -extreme_theta_threshold for theta in theta_array):
            indices_with_extreme_theta.append(idx)

        # Calculate cost
        theta_max = max(theta_array, key=abs)  # Maximum absolute theta value
        cost = a * theta_max + b * t_task
        costs.append({"index": idx, "cost": cost})

# Output results
print("Indices with extreme theta values:", indices_with_extreme_theta)
print("Costs for each dictionary:", costs)

## TODO: check the cost assignment
```

```python
# Find the lowest cost and its index
if costs:
    min_cost_entry = min(costs, key=lambda x: x["cost"])
    print("Lowest cost:", min_cost_entry["cost"])
    print("Index of lowest cost:", min_cost_entry["index"])
else:
    print("No costs calculated.")
```
